```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from io import BytesIO
import seaborn as sns
```

```python
url = "https://github.com/pritiyadav888/Housing-datasets/blob/main/canadian-us-residential-mortgage-arrears-foreclosure-rates-2002-2018-q1-en.xlsx?raw=true"
response = requests.get(url)

if response.status_code == 200:
    df = pd.read_excel(BytesIO(response.content))
else:
    print("Error downloading file")
```

```python
# Extract relevant data
data = df.iloc[[1, 2], [0] + list(range(1, 12))]

# Create a new DataFrame with proper column names
data.columns = ['Indicator', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012']

# Display the new DataFrame
data.head()
```

```python
# Melt the DataFrame into a long format
data_long = data.melt(id_vars=['Indicator'], var_name='Year', value_name='Rate')

# Display the transformed DataFrame
data_long.head()
```

```python
# Filter data for Canadian and U.S. mortgage arrears rates
canadian_arrears = data_long[data_long['Indicator'].str.contains('Canadian mortgage arrears')]
us_arrears = data_long[data_long['Indicator'].str.contains('U.S. prime fixed-rate mortgage arrears')]

# Reset the index
canadian_arrears = canadian_arrears.reset_index(drop=True)
us_arrears = us_arrears.reset_index(drop=True)
```

```python
# Merge the Canadian and U.S. mortgage arrears rates
merged_data = canadian_arrears.copy()
merged_data['US_Arrears_Rate'] = us_arrears['Rate']

# Rename the columns
merged_data = merged_data.rename(columns={'Rate': 'Canadian_Arrears_Rate', 'Year': 'Year'})

# Display the merged DataFrame
merged_data.head()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Canadian_Arrears_Rate', data=merged_data, label='Canadian Arrears Rate')
sns.lineplot(x='Year', y='US_Arrears_Rate', data=merged_data, label='US Arrears Rate')
plt.xlabel('Year')
plt.ylabel('Mortgage Arrears Rate')
plt.title('Mortgage Arrears Rate (2002-2012)')
plt.legend()
plt.show()
```

```python
correlation = merged_data['Canadian_Arrears_Rate'].corr(merged_data['US_Arrears_Rate'])
print(f"Correlation between Canadian and U.S. mortgage arrears rates: {correlation:.2f}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Canadian_Arrears_Rate', y='US_Arrears_Rate', data=merged_data)
plt.xlabel('Canadian Mortgage Arrears Rate')
plt.ylabel('US Mortgage Arrears Rate')
plt.title('Canadian vs. US Mortgage Arrears Rate (2002-2012)')
plt.show()
```

```python
# Convert 'Year' column to datetime format
merged_data['Year'] = pd.to_datetime(merged_data['Year'], format='%Y')

# Set 'Year' as index
merged_data.set_index('Year', inplace=True)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

canadian_decomposition = seasonal_decompose(merged_data['Canadian_Arrears_Rate'], model='additive', period=1)
us_decomposition = seasonal_decompose(merged_data['US_Arrears_Rate'], model='additive', period=1)

canadian_decomposition.plot()
plt.show()

us_decomposition.plot()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model for Canadian data
canadian_model = ARIMA(merged_data['Canadian_Arrears_Rate'], order=(1, 0, 0))
canadian_results = canadian_model.fit()

# Fit ARIMA model for US data
us_model = ARIMA(merged_data['US_Arrears_Rate'], order=(1, 0, 0))
us_results = us_model.fit()
```

```python
print("Canadian ARIMA Model Summary")
print(canadian_results.summary())

print("\nUS ARIMA Model Summary")
print(us_results.summary())
```

```python
# Forecast for the next 5 years
canadian_forecast = canadian_results.get_forecast(steps=5)
us_forecast = us_results.get_forecast(steps=5)

# Plot the forecasts
plt.figure(figsize=(12, 6))
merged_data['Canadian_Arrears_Rate'].plot(label='Canadian Arrears Rate')
merged_data['US_Arrears_Rate'].plot(label='US Arrears Rate')
canadian_forecast.predicted_mean.plot(label='Canadian Forecast')
us_forecast.predicted_mean.plot(label='US Forecast')
plt.xlabel('Year')
plt.ylabel('Mortgage Arrears Rate')
plt.title('Mortgage Arrears Rate Forecast (2002-2017)')
plt.legend()
plt.show()
```
